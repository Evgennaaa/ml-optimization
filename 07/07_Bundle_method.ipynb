Реализация Bundle-метода:

Написан код для реализации Bundle-метода, который хранит несколько субградиентов и использует их для улучшения сходимости.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
```

```python
class BundleMethodSVM:
    def __init__(self, learning_rate=0.01, lambda_param=0.1, max_iter=100):
        """
        Реализация Bundle-метода для SVM.
        :param learning_rate: скорость обучения
        :param lambda_param: параметр регуляризации
        :param max_iter: максимальное количество итераций
        """
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.max_iter = max_iter
        self.weights = None  
        self.subgradients = []  

    def hinge_loss_subgradient(self, X, y):
        """
        Вычисление субградиента функции потерь hinge loss.
        :param X: матрица признаков
        :param y: целевая переменная
        :return: субградиент
        """
        n_samples, n_features = X.shape
        subgradient = np.zeros(n_features)

        for i in range(n_samples):
            margin = y[i] * np.dot(X[i], self.weights)
            if margin < 1:  # Если ошибка или точка на границе
                subgradient -= y[i] * X[i]

        # Учитываем регуляризацию
        subgradient = subgradient / n_samples + self.lambda_param * self.weights
        return subgradient

    def fit(self, X, y):
        """
        Обучение модели с использованием Bundle-метода.
        :param X: матрица признаков
        :param y: целевая переменная
        """
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)  

        for iteration in range(self.max_iter):
            
            subgradient = self.hinge_loss_subgradient(X, y)
            self.subgradients.append(subgradient)

            # Вычисляем обновление весов через усреднение субградиентов
            avg_subgradient = np.mean(self.subgradients, axis=0)
            self.weights -= self.learning_rate * avg_subgradient

            if iteration % 10 == 0:
                loss = self.hinge_loss(X, y)
                print(f"Iteration {iteration}, Loss: {loss:.4f}")

    def hinge_loss(self, X, y):
        """
        Вычисление функции потерь hinge loss.
        :param X: матрица признаков
        :param y: целевая переменная
        :return: значение функции потерь
        """
        margins = 1 - y * (X @ self.weights)
        losses = np.maximum(0, margins)
        return np.mean(losses) + (self.lambda_param / 2) * np.dot(self.weights, self.weights)

    def predict(self, X):
        """
        Предсказания классов.
        :param X: матрица признаков
        :return: предсказания
        """
        return np.sign(X @ self.weights)
```

```python
# Реализация SVM с градиентным спуском (SGD)
class GradientDescentSVM:
    def __init__(self, learning_rate=0.01, lambda_param=0.1, max_iter=100):
        """
        Реализация SVM с градиентным спуском.
        :param learning_rate: скорость обучения
        :param lambda_param: параметр регуляризации
        :param max_iter: количество итераций
        """
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.max_iter = max_iter
        self.weights = None
        self.loss_history = []

    def hinge_loss(self, X, y):
        """
        Вычисление функции потерь hinge loss.
        """
        margins = 1 - y * (X @ self.weights)
        losses = np.maximum(0, margins)
        return np.mean(losses) + (self.lambda_param / 2) * np.dot(self.weights, self.weights)

    def fit(self, X, y):
        """
        Обучение модели с использованием градиентного спуска.
        """
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)

        for iteration in range(self.max_iter):
            margins = y * (X @ self.weights)
            gradient = np.zeros(n_features)

            for i in range(n_samples):
                if margins[i] < 1:
                    gradient -= y[i] * X[i]

            gradient = gradient / n_samples + self.lambda_param * self.weights
            self.weights -= self.learning_rate * gradient

            loss = self.hinge_loss(X, y)
            self.loss_history.append(loss)

            if iteration % 10 == 0:
                print(f"Gradient Descent - Iteration {iteration}, Loss: {loss:.4f}")

    def predict(self, X):
        """
        Предсказания классов.
        """
        return np.sign(X @ self.weights)
```

```python
df = pd.read_csv('German Credit Cards UCI.csv')



categorical_columns = ['Checking Account Status', 'Credit History', 'Purpose',
                       'Savings Account Balance', 'Present Employment Time (Years)',
                       'Personal Status and Gender', 'Other Debtors / Guarantors',
                       'Property', 'Other Credits', 'Home Ownership',
                       'Job Category', 'Telephone', 'Foreign Worker Status']


# Преобразуем целевую переменную: 1 = 1 (клиент надёжен), 2 = -1 (клиент ненадёжен)
df.rename(columns={"Loan Status": "Target"}, inplace=True)
df["Target"] = df["Target"].map({"fully paid": 1, "charged off": -1})

# Преобразуем категориальные признаки в one-hot encoding
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

```

```python

X = df.drop(columns=["Target"])
y = df["Target"].values


X_train, X_test, y_train, y_test = train_test_split(X.values, y, test_size=0.2, random_state=42)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
svm_real = BundleMethodSVM(learning_rate=0.05, lambda_param=0.1, max_iter=500)
svm_real.fit(X_train, y_train)

svm_gd = GradientDescentSVM(learning_rate=0.01, lambda_param=0.1, max_iter=500)
svm_gd.fit(X_train, y_train)


y_pred_real = svm_real.predict(X_test)
y_pred_gd = svm_gd.predict(X_test)


accuracy_real = np.mean(y_pred_real == y_test)
accuracy_gd = np.mean(y_pred_gd == y_test)



print(f"Точность SVM (Bundle Method): {accuracy_real:.2%}")
print(f"Точность SVM (Gradient Descent): {accuracy_gd:.2%}")

```

```python
plt.figure(figsize=(10, 6))

# График сходимости Bundle-метода
if hasattr(svm_real, "subgradients"):
    bundle_loss = [np.linalg.norm(g) for g in svm_real.subgradients]
    plt.plot(range(len(bundle_loss)), bundle_loss, label="Bundle Method Loss", linestyle="--")

# График сходимости градиентного спуска
if hasattr(svm_gd, "loss_history"):
    plt.plot(range(len(svm_gd.loss_history)), svm_gd.loss_history, label="Gradient Descent Loss")


plt.title("Сравнение сходимости функции потерь SVM (Bundle, GD, SMO)")
plt.xlabel("Итерация")
plt.ylabel("Hinge Loss")
plt.legend()
plt.grid(True)


plt.show()
```
