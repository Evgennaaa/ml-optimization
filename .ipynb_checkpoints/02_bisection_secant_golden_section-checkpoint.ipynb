# Задание 1

Реализуйте метод золотого сечения: Напишите программу для минимизации функции  с использованием метода золотого сечения.

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
def f(x):
  return (x - 2)**2

def golden_section_search(a, b, tol=1e-6):
    phi = (1 + 5**0.5) / 2

    c = b - (b - a)/phi
    d = a + (b - a)/phi

    f_c, f_d = f(c), f(d)  # значения функции
    iterations = 0


    while abs(b - a) > tol: # Продолжаем, пока длина интервала больше заданной точности
        iterations += 1

        if f_c < f_d:
          # Минимум находится в левом подинтервале [a, d].
            b, d, f_d = d, c, f_c
            c = b - (b - a)/phi
            f_c = f(c)
        else:
            a, c, f_c = c, d, f_d
            d = a + (b - a)/phi
            f_d = f(d)

    # Точка минимума — середина интервала
    x_min = (a + b) / 2
    return x_min, f(x_min), iterations


a = 0
b = 5
# Тест
x_min, f_min, iterations = golden_section_search(a, b)
print(f"Точка минимума: x = {x_min}")
print(f"Количество итераций: {iterations}")
```

```python
x = np.linspace(a, b, 500)
y = f(x)


plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x) = (x - 2)^2")
plt.scatter(x_min, f_min, color="red", label=f"Точка минимума ({x_min:.6f})")
plt.title("График функции f(x) = (x - 2)^2 с точкой минимума")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="gray", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

# Задание 2

Реализуйте метод Брента: Реализуйте гибридный метод Брента для функции

 image.png

Метод Брента – это гибридный численный метод, который сочетает в себе преимущества параболической интерполяции и метода деления интервала (метод золотого сечения).

```python
def f(x):
    return np.exp(x) + x**2


def brent_method(a, b, tol=1e-6):

    # x — текущая точка с минимальным значением функции,
    # w — предыдущая точка минимума,
    # v — точка перед w.
    x, w, v = a + (b - a) / 2, a + (b - a) / 2, a + (b - a) / 2
    fx, fw, fv = f(x), f(x), f(x)

    phi = (1 + 5**0.5) / 2
    d = 0  # Текущий шаг
    iterations = 0

    while abs(b - a) > tol:
        iterations += 1
        u = None # Новая точка для исследования

        # Параболическая интерполяция
        if x != w and x != v and w != v:
          # Вычисляем точку u
            u = x - ((x - w)**2 * (fx - fv) - (x - v)**2 * (fx - fw)) / \
                (2 * ((x - w) * (fx - fv) - (x - v) * (fx - fw)))

            # Проверяем, находится ли u внутри интервала и является ли шаг u достаточно малым
            if not (a + tol <= u <= b - tol) or abs(u - x) > abs(b - a) / 2:
                u = None

        # Если параболическая интерполяция не удалась, используем золотое сечение
        if u is None:
            if x < (b + a) / 2:
                u = x + (b - x) / phi
            else:
                u = x - (x - a) / phi

        # Обновляем интервал и точки
        fu = f(u)
        if fu <= fx: # Если значение функции в u меньше текущего минимума fx
            if u < x:
                b = x # Сужаем интервал справа
            else:
                a = x
             # Обновляем точки
            v, fv = w, fw
            w, fw = x, fx
            x, fx = u, fu
        else:
            if u < x:
                a = u
            else:
                b = u
            if fu <= fw or w == x:
                v, fv = w, fw
                w, fw = u, fu
            elif fu <= fv or v == x or v == w:
                v, fv = u, fu

    return x, fx, iterations


a, b = -3, 2
x_min, f_min, iterations = brent_method(a, b)

print(f"Точка минимума: x = {x_min}")
print(f"Количество итераций: {iterations}")

```

```python

x = np.linspace(a, b, 500)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x) = e^x + x^2")
plt.scatter(x_min, f_min, color="red", label=f"Точка минимума ({x_min:.6f}, {f_min:.6f})")
plt.title("Метод Брента: График функции f(x) = e^x + x^2 с точкой минимума")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axvline(x_min, color="gray", linestyle="--", label=f"x_min = {x_min:.6f}")
plt.legend()
plt.grid()
plt.show()
```

# Задание 3

Реализуйте метод бисекции: Напишите программу для нахождения корня функции методом бисекции.

image.png

Метод бисекции (или метод деления отрезка пополам) — это численный метод нахождения корней функции или минимизации функции. Он основан на итеративном делении заданного интервала пополам до тех пор, пока длина интервала не станет меньше заданной точности

```python
def func(x):
    return x**3 - 3*x + 1


def bisection(a, b, min_value, max_iterations=1000):
    if func(a) * func(b) >= 0:
        raise ValueError("Функция должна иметь разные знаки на концах отрезка [a, b].")

    iterations = 0

    while abs(b - a) > min_value and iterations < max_iterations:
        c = (a + b) / 2 # Середина интервала
        f_c = func(c)

        if f_c == 0:  # Точный корень найден
            return c, f_c, iterations
        elif func(a) * f_c < 0:  # Корень в левой половине
            b = c
        else:  # Корень в правой половине
            a = c

        iterations += 1

    # Приближённый корень
    c = (a + b) / 2
    return c, func(c), iterations

# Задаём интервал и параметры
a = -2
b = 2
min_value = 0.001  # Точность

# Выполняем вычисление корня
result, f_value, iterations = bisection(a, b, min_value)

print("Приближённый корень:", result)
print("Значение функции в корне:", f_value)
print("Количество итераций:", iterations)
```

# Задание 4

Реализуйте метод секущей: Напишите программу для нахождения корня функции
методом секущей.

image.png

Метод секущих — это численный метод нахождения корня уравнения  f(x) = 0 . Он является модификацией метода Ньютона, где вместо вычисления производной используется приближённое значение, полученное с помощью двух последних точек.

```python
import numpy as np
import matplotlib.pyplot as plt


def func(x):
    return x**2 - 4


def secant_method(x0, x1, tol=1e-6, max_iter=100):

    points = [(x0, func(x0)), (x1, func(x1))]
    for i in range(max_iter):
        f_x0, f_x1 = func(x0), func(x1)

        # Проверка деления на ноль
        if f_x1 - f_x0 == 0:
            raise ValueError("Деление на ноль: f(x1) - f(x0) = 0.")

        # Формула метода секущих для нахождения следующей точки
        x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)
        points.append((x2, func(x2)))

        # Условие остановки: если значение функции в новой точке меньше заданной точности
        if abs(func(x2)) < tol:
            return x2, func(x2), i + 1, points

        # Обновляем точки
        x0, x1 = x1, x2

    # Если не сошлось
    raise ValueError("Метод не сошёлся за максимальное число итераций.")


x0, x1 = 0, 3  # Начальные приближения


root, f_value, iterations, points = secant_method(x0, x1)


print(f"Приближённый корень: x = {root}")
print(f"Значение функции в корне: f(x) = {f_value}")
print(f"Количество итераций: {iterations}")
```

```python
x_vals = np.linspace(-3, 3, 500)
y_vals = func(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="f(x) = x^2 - 4", color="blue")
plt.axhline(0, color="black", linestyle="--", linewidth=0.7)
plt.scatter(*zip(*points), color="red", label="Точки итераций", zorder=5)
plt.scatter(root, f_value, color="green", label=f"Корень: x = {root:.4f}", zorder=10)
plt.title("Метод секущих для функции f(x) = x^2 - 4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()
plt.show()
```
