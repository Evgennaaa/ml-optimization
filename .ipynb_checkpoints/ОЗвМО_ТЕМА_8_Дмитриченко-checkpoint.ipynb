Стохастический градиентный спуск

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
class LinearRegressionGD:
    def __init__(self, lr=0.01, max_iter=1000, tol=1e-3, intercept = True):
        # скорость обучения градиентного спуска
        self.learning_rate = lr
        # максимальное число итераций
        self.max_iteration = max_iter
        # критерий сходимости
        self.tolerance_convergence  = tol
        # наличие свободного члена
        self.intercept = intercept
        # инициальзация весов: None
        self.theta = None
        self.n = None
        self.d = None

    def fit(self, X, y):
        self.X = X.copy()
        if self.intercept:
            self.X = np.hstack((np.ones((self.X.shape[0], 1)), self.X))
        # размерность вектора X
        self.n, self.d = self.X.shape
        # инициализация случайных весов
        self.theta = np.random.randn(self.d)
        steps, errors = [], []
        step = 0
        for _ in range(self.max_iteration):
            grad = self.gradient(self.X, y)
            self.theta -= self.learning_rate * grad
            new_error = ((y - self.predict(X)).T @ (y - self.predict(X))) / self.n
            step += 1
            steps.append(step)
            errors.append(new_error)
            # проверка сходимости
            if np.linalg.norm(grad) < self.tolerance_convergence:
                break
        return steps, errors

    def gradient(self, X, y):
        return X.T @ (X @ self.theta - y) / len(y)

    def predict(self, X):
        if self.intercept:
            X_ = np.hstack((np.ones((X.shape[0],1)), X))
        else:
            X_ = X
        return X_ @ self.theta

    def MSE(self, X, y):
        return ((y - self.predict(X)).T @ (y - self.predict(X))) / len(y)

    def MAE(self, X, y):
        return abs(y - self.predict(X)).mean()

    def MAPE(self, X, y):
        return abs((y - self.predict(X))/y).mean()
```

```python
X = np.random.randn(100, 5)
y = X @ np.array([1, 2, 3, 4, 5]) + np.random.randn(100) * 2 + 6

start_time = time.time()
modelGD = LinearRegressionGD(lr=0.01, max_iter=400, tol=1e-3, intercept = True)
steps, errors = modelGD.fit(X, y)
print("Время обучения:", time.time() - start_time)
# Predict using predict method from model
y_pred = modelGD.predict(X)
print("MSE: ", modelGD.MSE(X, y))
print("Коэффициенты: ", modelGD.theta)
plt.scatter(X[:,2], y)
plt.scatter(X[:,2], y_pred)
```

```python
class LinearRegressionSGD:
    def __init__(self, lr=0.01, max_iter=1000, batch_size=32, tol=1e-3, intercept = True):
        # скорость обучения градиентного спуска
        self.learning_rate = lr
        # максимальное число итераций
        self.max_iteration = max_iter
        # размер мини-батча
        self.batch_size = batch_size
        # критерий сходимости
        self.tolerance_convergence  = tol
        # наличие свободного члена
        self.intercept = intercept
        # инициальзация весов: None
        self.theta = None
        self.n = None
        self.d = None

    def fit(self, X, y):
        self.X = X.copy()
        self.y = y.copy()
        if self.intercept:
            self.X = np.hstack((np.ones((self.X.shape[0], 1)), self.X))
        # размерность вектора X
        self.n, self.d = self.X.shape
        # инициализация случайных весов
        self.theta = np.random.randn(self.d)
        steps, errors = [], []
        step = 0
        for _ in range(self.max_iteration):
            # перемешивание данных
            indices = np.random.permutation(self.n)
            X = self.X[indices]
            y = self.y[indices]
            # цикл по мини-батчам
            for i in range(0, self.n, self.batch_size):
                X_batch = X[i:i+self.batch_size]
                y_batch = y[i:i+self.batch_size]
                grad = self.gradient(X_batch, y_batch)
                self.theta -= self.learning_rate * grad
            new_error = ((self.y - self.X @ self.theta).T @ (self.y - self.X @ self.theta)) / self.n
            step += 1
            steps.append(step)
            errors.append(new_error)
            # проверка сходимости
            if np.linalg.norm(grad) < self.tolerance_convergence:
                break
        return steps, errors

    def gradient(self, X, y):
        return X.T @ (X @ self.theta - y) / len(y)

    def predict(self, X):
        if self.intercept:
            X_ = np.hstack((np.ones((X.shape[0],1)), X))
        else:
            X_ = X
        return X_ @ self.theta

    def MSE(self, X, y):
        return ((y - self.predict(X)).T @ (y - self.predict(X))) / len(y)

    def MAE(self, X, y):
        return abs(y - self.predict(X)).mean()

    def MAPE(self, X, y):
        return abs((y - self.predict(X))/y).mean()
```

```python
def train_test_split_(x, y, test_size=0.2, random_state=None):
    data = np.concatenate((x, y.reshape(y.shape[0],1)), axis=1)
    # Задаем seed для воспроизводимости результатов, если указан random_state
    if random_state is not None:
        np.random.seed(random_state)

    # Вычисляем количество элементов для тестовой выборки
    num_test = int(len(data) * test_size)

    # Случайно перемешиваем индексы
    shuffled_indices = np.random.permutation(len(data))

    # Выбираем индексы для тестовой выборки
    test_indices = shuffled_indices[:num_test]

    # Выбираем индексы для обучающей выборки
    train_indices = shuffled_indices[num_test:]

    # Разделяем данные на обучающую и тестовую выборки
    train_data = data[train_indices]
    test_data = data[test_indices]

    return train_data[:,:-1], train_data[:,-1], test_data[:,:-1], test_data[:,-1]
```

Реализуем L2-регуляризацию с помощью прямых вычислений:
\begin{align}w &= (X^TX + \lambda I_d)^{-1}X^Ty\end{align}

```python
train_scores, test_scores = [], []
id = np.eye(x_train.shape[1]+1)
id[0][0] = 0
print("l, MSE_test - MSE_train:")
for l in np.logspace(-20, 20, 100):
    w = np.linalg.inv(x_train1.T @ x_train1 + l*id) @ x_train1.T @ y_train
    MSE_train = (y_train - x_train1 @ w).T @ (y_train - x_train1 @ w)/ x_train1.shape[0]
    train_scores.append(MSE_train)
    MSE_test = (y_test - x_test1 @ w).T @ (y_test - x_test1 @ w)/ x_test1.shape[0]
    test_scores.append(MSE_test)
    print(l, MSE_test - MSE_train)
plt.plot(train_scores, 'g')
plt.plot(test_scores, 'r')

start_time = time.time()
w = np.linalg.inv(x_train1.T @ x_train1 + 0.038*id) @ x_train1.T @ y_train
MSE_train = (y_train - x_train1 @ w).T @ (y_train - x_train1 @ w)/ x_train1.shape[0]
MSE_test = (y_test - x_test1 @ w).T @ (y_test - x_test1 @ w)/ x_test1.shape[0]
print("Время обучения:", time.time() - start_time)
print("MSE_train, MSE_test", MSE_train, MSE_test)
print("Коэффициенты: ", w)
print("np.linalg.det(x_train.T @ x_train): ", np.linalg.det(x_train1.T @ x_train1))
print("np.linalg.det(x_train.T @ x_train + 0.43287*id): ", np.linalg.det(x_train1.T @ x_train1 + 0.43287*id))
```

Объясните, почему коэффициенты всех построенных моделей сильно отличаются между собой.

Основные причины различий в коэффициентах моделей:
	1.	Разные методы оптимизации (GD vs. SGD)
	•	Градиентный спуск (GD) обновляет коэффициенты, используя всю выборку данных, что даёт более стабильные результаты.
	•	Стохастический градиентный спуск (SGD) обновляет коэффициенты по одному случайному примеру за итерацию, из-за чего веса могут колебаться и сходиться к разным значениям.
	2.	Случайная инициализация весов
	•	Оба метода начинают с разных случайных значений коэффициентов, что влияет на их финальные значения.
	3.	Чувствительность к параметрам обучения
	•	Скорость обучения (lr) влияет на то, насколько сильно изменяются коэффициенты в каждой итерации.
	•	Критерий сходимости (tol) может остановить обучение на разных этапах, приводя к разным итоговым коэффициентам.
	4.	Шум в данных
	•	В выборке присутствует случайный шум (np.random.randn(100) * 2), что влияет на процесс оптимизации, особенно для SGD.
	5.	Разное количество итераций до сходимости
	•	В GD корректировки коэффициентов происходят медленнее, но стабильнее.
	•	В SGD шаги резче, но могут приводить к флуктуациям (скачкам).


Подберите в цикле оптимальное, с Вашей точки зрения, значение размера мини-батча и поясните Ваш выбор.

```python
class LinearRegressionMBGD:
    """Мини-батч градиентный спуск"""
    def __init__(self, lr=0.01, max_iter=1000, tol=1e-3, batch_size=32, intercept=True):
        self.learning_rate = lr
        self.max_iteration = max_iter
        self.tolerance_convergence = tol
        self.batch_size = batch_size
        self.intercept = intercept
        self.theta = None

    def fit(self, X, y):
        if self.intercept:
            X = np.hstack((np.ones((X.shape[0], 1)), X))

        self.n, self.d = X.shape
        self.theta = np.random.randn(self.d)

        for _ in range(self.max_iteration):
            batch_indices = np.random.choice(self.n, self.batch_size, replace=False)  # Выбираем случайный мини-батч
            X_batch = X[batch_indices]
            y_batch = y[batch_indices]

            grad = X_batch.T @ (X_batch @ self.theta - y_batch) / self.batch_size
            self.theta -= self.learning_rate * grad

            if np.linalg.norm(grad) < self.tolerance_convergence:
                break

    def predict(self, X):
        if self.intercept:
            X = np.hstack((np.ones((X.shape[0], 1)), X))
        return X @ self.theta

    def mse(self, X, y):
        return ((y - self.predict(X)) ** 2).mean()


# Перебор различных размеров мини-батчей
batch_sizes = [1, 8, 16, 32, 64, 128, 256, 404]  # От стохастического до полного градиентного спуска
results = []

for batch_size in batch_sizes:
    start_time = time.time()
    model_mbgd = LinearRegressionMBGD(lr=0.01, max_iter=1000, batch_size=batch_size)
    model_mbgd.fit(X_train, y_train)
    training_time = time.time() - start_time
    test_mse = model_mbgd.mse(X_test, y_test)

    results.append([batch_size, training_time, test_mse])

# Создаем DataFrame с результатами
batch_comparison_df = pd.DataFrame(results, columns=["Размер мини-батча", "Время обучения (сек)", "MSE на тесте"])

# Отображение результатов
print(batch_comparison_df)
```

Основные выводы:
	•	Мини-батч размером 1 (полностью стохастический градиентный спуск) дает крайне высокое MSE, что указывает на нестабильность обучения.
	•	Мини-батчи от 8 до 64 показывают хорошие результаты, причем размер 32 дает лучшее соотношение точности и скорости.
	•	Размер 64 также работает стабильно, но немного увеличивает MSE.

Реализуйте в классе LinearRegressionSGD стохастический градиентный спуск с регуляризацией L1, L2, ElasticNet и подберите оптимальные параметры регуляризации, чтобы сократить разрыв качества на обучающей и тестовой выборке.

```python
import numpy as np
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

class LinearRegressionSGD:
    """
    Стохастический градиентный спуск с поддержкой L1, L2 и ElasticNet регуляризации.
    """
    def __init__(self, lr=0.01, max_iter=1000, tol=1e-3, intercept=True,
                 l1_ratio=0.0, l2_ratio=0.0, elasticnet_ratio=0.5):
        self.learning_rate = lr
        self.max_iteration = max_iter
        self.tolerance_convergence = tol
        self.intercept = intercept
        self.l1_ratio = l1_ratio  # Коэффициент L1 регуляризации
        self.l2_ratio = l2_ratio  # Коэффициент L2 регуляризации
        self.elasticnet_ratio = elasticnet_ratio  # Вес L1 в ElasticNet
        self.theta = None

    def fit(self, X, y):
        if self.intercept:
            X = np.hstack((np.ones((X.shape[0], 1)), X))

        self.n, self.d = X.shape
        self.theta = np.random.randn(self.d)

        for _ in range(self.max_iteration):
            i = np.random.randint(self.n)  # Выбираем случайный объект
            xi = X[i, :].reshape(1, -1)
            yi = y[i]
            grad = xi.T @ (xi @ self.theta - yi)

            # Добавляем L1, L2, ElasticNet регуляризацию
            if self.l1_ratio > 0:
                grad += self.l1_ratio * np.sign(self.theta)
            if self.l2_ratio > 0:
                grad += self.l2_ratio * self.theta
            if self.l1_ratio > 0 and self.l2_ratio > 0:  # ElasticNet
                grad += self.elasticnet_ratio * self.l1_ratio * np.sign(self.theta) + \
                        (1 - self.elasticnet_ratio) * self.l2_ratio * self.theta

            self.theta -= self.learning_rate * grad.flatten()

            if np.linalg.norm(grad) < self.tolerance_convergence:
                break

    def predict(self, X):
        if self.intercept:
            X = np.hstack((np.ones((X.shape[0], 1)), X))
        return X @ self.theta

    def mse(self, X, y):
        return ((y - self.predict(X)) ** 2).mean()

    def tune_regularization(self, X_train, y_train, X_test, y_test, l1_values, l2_values, elasticnet_values):
        best_params = None
        best_mse_gap = float('inf')

        for l1 in l1_values:
            for l2 in l2_values:
                for en_ratio in elasticnet_values:
                    self.l1_ratio = l1
                    self.l2_ratio = l2
                    self.elasticnet_ratio = en_ratio
                    self.fit(X_train, y_train)
                    train_mse = self.mse(X_train, y_train)
                    test_mse = self.mse(X_test, y_test)
                    mse_gap = abs(train_mse - test_mse)

                    if mse_gap < best_mse_gap:
                        best_mse_gap = mse_gap
                        best_params = (l1, l2, en_ratio, train_mse, test_mse)

        return best_params
```

```python
# Подбор оптимальных параметров регуляризации
l1_values = [0.001, 0.01, 0.1, 1.0]
l2_values = [0.001, 0.01, 0.1, 1.0]
elasticnet_values = [0.1, 0.5, 0.9]

# Создаем объект модели
model = LinearRegressionSGD(lr=0.01, max_iter=1000, tol=1e-3, intercept=True)

# Подбор оптимальных параметров регуляризации
best_params = model.tune_regularization(X_train, y_train, X_test, y_test, l1_values, l2_values, elasticnet_values)
print("Лучшие параметры регуляризации:", best_params)
```

```python
train_loss = model.mse(X_train, y_train)
test_loss = model.mse(X_test, y_test)

# Вывод
print(f"Train Loss (MSE): {train_loss:.4f}")
print(f"Test Loss (MSE):  {test_loss:.4f}")
```

1.	L1-регуляризация (l1_ratio = 1.0)
	•	Максимальное значение из тестируемых, что говорит о том, что модель предпочитает Lasso-регуляризацию.
	•	Это означает, что многие коэффициенты стали равны нулю, что упрощает модель и делает её более устойчивой к переобучению.
	2.	L2-регуляризация (l2_ratio = 0.001)
	•	Очень маленькое значение, что указывает на то, что Ridge-регуляризация практически не влияет.
	•	Это подтверждает, что модель больше ориентируется на L1-регуляризацию.
	3.	ElasticNet (elasticnet_ratio = 0.9)
