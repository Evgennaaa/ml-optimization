
Реализация классического градиентного спуска для минимизации функции.


image.png

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x[0]**2 + 2 * x[1]**2

def gradient_f(x):
    return np.array([2 * x[0], 4 * x[1]])


def gradient_descent_adaptive(x0, alpha, beta, threshold, max_iter):
    """
    x0: Начальная точка (список или numpy.array)
    alpha: Начальный шаг
    beta: Коэффициент уменьшения шага
    threshold: Порог для критерия остановки
    max_iter: Максимальное количество итераций
    """
    # Преобразовние начальной точки в массив
    x = np.array(x0, dtype=float)
    trajectory = [x.copy()]  # Для хранения траектории точек

    for i in range(max_iter):
        grad = gradient_f(x)
        grad_norm = np.linalg.norm(grad)  # Норма градиента для проверки остановки

        if grad_norm < threshold:  # проверка критерия остановки
            print(f"Метод сошелся за {i + 1} итераций.")
            break

        # Адаптивный поиск шага
        step = alpha
        while f(x - step * grad) >= f(x): # Уменьшаем шаг, пока новая точка не улучшает значение функции
            step *= beta # Уменьшаем шаг

        x = x - step * grad  # Обновляем точку
        trajectory.append(x.copy())  # Сохраняем точку

    return np.array(trajectory), i + 1, x


x0 = [2, 2]
alpha = 1
beta = 0.5
threshold = 1e-6
max_iter = 100


trajectory, iterations, minimum_point = gradient_descent_adaptive(x0, alpha, beta, threshold, max_iter)


print(f"Точка минимума: {minimum_point}")
```

```python

x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = X1**2 + 2 * X2**2

plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Z, levels=50, cmap="viridis")
trajectory = np.array(trajectory)
plt.plot(trajectory[:, 0], trajectory[:, 1], marker="o", color="red", label="Траектория сходимости")
plt.scatter(minimum_point[0], minimum_point[1], color="blue", s=100, label=f"Точка минимума: ({minimum_point[0]:.4f}, {minimum_point[1]:.4f})")
plt.title("Градиентный спуск с адаптивным шагом")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.grid()
plt.show()
```

Реализация метода Ньютона для функции.


image.png

```python
def f(x):
    return x[0]**2 + 2 * x[1]**2

def gradient_f(x):
    return np.array([2 * x[0], 4 * x[1]])

def hessian_f(x):
    return np.array([[2, 0], [0, 4]])


def newton_method(x0, threshold=1e-6, max_iter=100, gamma=1e-3):
    x = np.array(x0, dtype=float)
    trajectory = [x.copy()]  # Для хранения траектории точек

    for i in range(max_iter):
        grad = gradient_f(x)
        hessian = hessian_f(x)

        # Проверка положительной определенности гессиана
        if not np.all(np.linalg.eigvals(hessian) > 0):
          hessian += gamma * np.eye(len(x))

        # Шаг Ньютона: H^(-1) * (-grad)
        delta_x = np.linalg.solve(hessian, -grad)
        x = x + delta_x  # Обновляем точку
        trajectory.append(x.copy())  # Сохраняем точку

        # Критерий остановки
        if np.linalg.norm(grad) < threshold:
            print(f"Метод Ньютона сошелся за {i + 1} итераций.")
            return np.array(trajectory), i + 1

    print("Достигнуто максимальное количество итераций.")
    return np.array(trajectory), max_iter


x0 = [2, 2]
trajectory_newton, iterations_newton = newton_method(x0)


print(f"Точка минимума: {trajectory_newton[-1]}")

```

```python
plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Z, levels=50, cmap="viridis")
plt.plot(trajectory[:, 0], trajectory[:, 1], marker="o", color="green", label="Градиентный спуск")
plt.plot(trajectory_newton[:, 0], trajectory_newton[:, 1], marker="o", color="red", label="Метод Ньютона")
plt.scatter(0, 0, color="blue", s=100, label="Точка минимума")
plt.title("Сравнение: Градиентный спуск и Метод Ньютона")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.grid()
plt.show()
```
